# Weather Mapping Playground

```python
import numpy as np
import earthkit.plots.quickmap as qmap
import earthkit as ek
```

```python
ds = ek.data.from_source("cds",
        "reanalysis-era5-single-levels",
        variable=["2m_temperature"],
        product_type="reanalysis",
        # area=[50, -75, 40, -65], #N,W,S,E
        date="2012-05-10",
        time="12:00")
```

```python
import matplotlib.pyplot as plt


plt.imshow(ds.to_array().squeeze())
```

```python
new_england_cood = [-78, -65, 40, 48]
new_england_domain = ek.plots.geo.domains.Domain(
    bbox=new_england_cood,
    name="New England",
)


fig = ek.plots.Figure()
fig.subplot_titles("{domain}")

map = fig.add_map(domain=new_england_domain) #E, W, N, S
map.coastlines()
map.borders()
map.land()
map.gridlines()
map.administrative_areas()
map.cities(adjust_labels=True)

map.plot(ds, units="fahrenheit")


#fig.legend(vmin=30, vmax=40)


fig.show()
```
